```python
import pandas as pd

# Load the training dataset
training_data = pd.read_csv('/kaggle/input/twitter-entity-sentiment-analysis/twitter_training.csv')

# Display the first few rows of the dataset
training_data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count the occurrences of each sentiment
sentiment_counts = training_data['Positive'].value_counts()

# Create a bar plot for sentiment distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette='viridis')
plt.title('Sentiment Distribution in Social Media Data')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()

# Save the plot as an HTML file
plt.savefig('sentiment_distribution.png')
plt.show()
```

```python
import pandas as pd

# Read the CSV file without headers
df = pd.read_csv('/kaggle/input/twitter-entity-sentiment-analysis/twitter_validation.csv', header=None)

# Display the first few rows
print(df.head())

# Display column information
print(df.info())
```

```python
# Assigning appropriate column names
column_names = ['ID', 'Platform', 'Sentiment', 'Text']
df.columns = column_names

# Display the updated dataframe structure
print(df.head())
print(df.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style='whitegrid')

# Count the occurrences of each sentiment by platform
sentiment_distribution = df.groupby(['Platform', 'Sentiment']).size().unstack(fill_value=0)

# Plotting the sentiment distribution across platforms
plt.figure(figsize=(12, 6))
sentiment_distribution.plot(kind='bar', stacked=True, ax=plt.gca())
plt.title('Sentiment Distribution Across Platforms')
plt.xlabel('Platform')
plt.ylabel('Number of Mentions')
plt.xticks(rotation=45)
plt.legend(title='Sentiment')
plt.tight_layout()

# Save the plot as an HTML file
plt.savefig('sentiment_distribution.png')
plt.show()
```

```python
# Analyzing the relationship between platforms and sentiment

# Create a count of sentiments by platform
platform_sentiment_counts = df.groupby(['Platform', 'Sentiment']).size().unstack(fill_value=0)

# Plotting the relationship between platforms and sentiment
plt.figure(figsize=(12, 6))
sns.heatmap(platform_sentiment_counts, annot=True, fmt='d', cmap='YlGnBu', cbar_kws={'label': 'Number of Mentions'})
plt.title('Relationship Between Platforms and Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Platform')
plt.tight_layout()

# Save the plot as an HTML file
plt.savefig('platform_sentiment_relationship.png')
plt.show()
```

```python
# Cleaning and restructuring the data

# Check for missing values and duplicates
missing_values = df.isnull().sum()
duplicates = df.duplicated().sum()

# Drop duplicates
cleaned_df = df.drop_duplicates()

# Reset index after dropping duplicates
cleaned_df.reset_index(drop=True, inplace=True)

# Display the results of cleaning
missing_values, duplicates, cleaned_df.head()
```
